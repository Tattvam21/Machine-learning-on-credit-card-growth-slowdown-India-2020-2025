## Checking for outliers

```python
from scipy import stats
import numpy as np
import pandas as pd
# SELECT ONLY NUMERICAL COLUMN
numerical_columns = df_credit_card_growth_slowdown_india_2020_2025.select_dtypes(include=[np.number]).columns
print(f"🔍 Checking for outliers in the following columns: {list(numerical_columns)}\n")
# COMPUTE Z-SCORE
z_scores = np.abs(stats.zscore(df_credit_card_growth_slowdown_india_2020_2025[numerical_columns]))
# DATA FRAME TO STORE OUTLIER'S DETAIL
outlier_rows = pd.DataFrame()
outlier_summary = {}  # DICTIONARY TO TRACK WHICH COLUMN HAVE OUTLIER'S
# ITRATE THROUGH NUMERICAL COLUMNS
for column in numerical_columns:
    #FIND OUTLIER'S INDICES WHERE Z-SCORE > 3
    outlier_indices = df_credit_card_growth_slowdown_india_2020_2025.index[z_scores[column] > 3].tolist()
    # STORE THE OUTLIER'S STATUS IN THE SUMMARY DICTIONARY
    outlier_summary[column] = len(outlier_indices) > 0
    # PRINT OUTLIER'S DETAIL IF FOUND
    if outlier_indices:
        print(f"🔹 **Column: {column}** - Outliers detected in {len(outlier_indices)} rows.")
        outlier_rows = pd.concat([outlier_rows, df_credit_card_growth_slowdown_india_2020_2025.loc[outlier_indices]], axis=0)

# REMOVE DUPLICATE ROW
outlier_rows = outlier_rows.drop_duplicates()
# DISPLAY TOTAL COUNT OF OUTLIER'S
print(f"\n✅ **Total outliers detected: {outlier_rows.shape[0]}**\n")
#PRINT SUMMARY OF COLUMNS CHECKED AND WHETHER OUTLIER'S WERE FOUND
for col, has_outliers in outlier_summary.items():
    status = "✅ No outliers found" if not has_outliers else "❌ Outliers detected"
    print(f"📊 Column: {col} -> {status}")
# DISPLAY FIRST FEW OUTLIER'S ROW IF IT EXIST
if not outlier_rows.empty:
    print("\n🔍 Sample of detected outliers:")
    print(outlier_rows.head())
else:
    print("\n🎉 No outliers detected in the dataset!")
```

```python
# Create a copy of the original dataset
df_cleaned = df_credit_card_growth_slowdown_india_2020_2025.copy()

# Function to cap outliers using IQR
def cap_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    df[column] = df[column].apply(
        lambda x: lower_bound if x < lower_bound else upper_bound if x > upper_bound else x
    )
    return df

# Columns to apply IQR outlier handling
columns_to_fix = ['Monthly_Spend', 'Outstanding_Balance']

# Apply IQR capping to the copy
for col in columns_to_fix:
    df_cleaned = cap_outliers_iqr(df_cleaned, col)
```

```python
from sklearn.linear_model import Lasso
from sklearn.feature_selection import RFE
from sklearn.metrics import mean_squared_error, r2_score

# Initialize the Lasso Regressor
lasso = Lasso(alpha=1.0)  # You can tune the alpha value

# Initialize RFE with Lasso
# Choose the number of features to select (e.g., 10 or any value ≤ total features)
n_features_to_select = 100
rfe = RFE(estimator=lasso, n_features_to_select=n_features_to_select)

# Fit RFE on training data
rfe.fit(X_train, y_train)

# Get the selected features
X_train_rfe = rfe.transform(X_train)
X_test_rfe = rfe.transform(X_test)

# Fit Lasso on selected features
lasso.fit(X_train_rfe, y_train)

# Predict
y_pred = lasso.predict(X_test_rfe)

# Evaluate
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Selected Features Mask:", rfe.support_)
print("Feature Ranking:", rfe.ranking_)
print("Mean Squared Error:", mse)
print("R² Score:", r2)
```
